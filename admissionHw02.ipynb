**<font size=3 face='STZhongSong'>三、數據建模、訓練、測試、指標比較</font>** 

```python
import statsmodels.formula.api as smf  #第一組採用多元回歸
```

```python
df.columns=['GREScore','TOEFLScore','UniversityRating','SOP','LOR','CGPA','Research','chanceofAdmit']
```

```python
df.sample(5)
```

```python
model_1 = smf.ols('chanceofAdmit~GREScore+TOEFLScore+UniversityRating+SOP+LOR+CGPA+Research', df).fit()
model_1.summary()   #多元回歸第一種組合
```

```python
model_2 = smf.ols('chanceofAdmit~GREScore+TOEFLScore+UniversityRating+SOP+LOR+CGPA+Research+GREScore*TOEFLScore*CGPA', df).fit()
model_2.summary()  #多元回歸第二種組合，GRE,TOELF,CGPA都是成績有關
```

```python
#model = smf.ols('chanceofAdmit~UniversityRating+SOP+LOR+Research+GREScore*TOEFLScore*CGPA', df).fit()
#model.summary()  #same as above
```

```python
x = df.loc[:,df.columns != 'chanceofAdmit']
y = df['chanceofAdmit']
x.sample(5)
```

```python
y.sample(5)
```

```python
y1_hat = model_1.predict(x)  #多元回歸第一種組合預測
y1_hat.head(5)
```

```python
model1_RMSE = np.sqrt(np.mean(np.square(y1_hat-y)))   #RMSE指標 結果一
model1_RMSE
```

```python
y2_hat = model_2.predict(x)  #多元回歸第二種組合預測
y2_hat.head(5)
```

```python
model2_RMSE = np.sqrt(np.mean(np.square(y2_hat-y)))   #RMSE指標 結果二
model2_RMSE
```

```python
from sklearn.linear_model import Ridge,Lasso
#引入岭回归模型、Lasso
```

```python
ridge=Ridge().fit(x,y)      #第二組採用岭回归模型
```

```python
y3_hat=ridge.predict(x)     #岭回归模型預測
y3_hat.shape
```

```python
model3_RMSE = np.sqrt(np.mean(np.square(y3_hat-y)))  #RMSE指標 結果三
model3_RMSE
```

```python
lasso=Lasso().fit(x,y)     #第三組採用lasso模型
```

```python
y4_hat=lasso.predict(x)    #lasso模型預測
y4_hat.shape
```

```python
model4_RMSE = np.sqrt(np.mean(np.square(y4_hat-y))) #RMSE指標 結果四
model4_RMSE
```

**<font size=3 face='STZhongSong'>四、總結，用多元回歸和岭回歸效果都不錯，尤其是多元回歸搭配有交互項，
    代表資料庫中挑選出的自變量對畢業許可，指標性很高</font>** 

**<font size=3 face='STZhongSong'>五、額外測試，想說把SOP等級拆開，分別當自變量看對分析的影響，因為P值SOP是最高的，結果沒太大改變</font>**
