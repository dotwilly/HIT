```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from zipfile import ZipFile
```

```python
with ZipFile('weather-dataset-rattle-package.zip', 'r') as z:  # 访问zip文件
    f = z.open('weatherAUS.csv')    # 选择zip文件里的数据
    df = pd.read_csv(f,index_col=0)    # 读取数据

df.shape  
```

```python
df.describe() #查看資料狀況
```

```python
df.isnull().sum() #查看缺失值
```

```python
df = df.drop(columns=['Location','RISK_MM']) #無關和會降低預測準度的變量刪除
df.head(5)
```

**<font size=3 face='STZhongSong'>三、缺失值處理</font>** 

```python
df.isnull().sum() #查看缺失值
```

```python
df = df.drop(columns=['Evaporation','Sunshine','WindDir9am','Pressure9am','Pressure3pm','Cloud9am','Cloud3pm']) #超過萬筆資料缺失刪除該變量影響
df.head(5)
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
mean_MinTemp = df['MinTemp'].mean()#溫度、濕度、降雨量相關採用平均值填充，因為觀察其數據狀況得出可以如此
mean_MinTemp
mean_MaxTemp = df['MaxTemp'].mean()
mean_MaxTemp
mean_Temp9am = df['Temp9am'].mean()
mean_Temp9am
mean_Temp3pm = df['Temp3pm'].mean()
mean_Temp3pm
mean_Rainfall = df['Rainfall'].mean()
mean_Rainfall
mean_Humidity9am = df['Humidity9am'].mean()
mean_Humidity9am
mean_Humidity3pm = df['Humidity3pm'].mean()
mean_Humidity3pm
```

```python
tmp = df['MinTemp'].copy()
tmp[df.MinTemp.isnull()] = mean_MinTemp
df['MinTemp_fill'] = tmp
del tmp
df = df.drop(['MinTemp'],axis=1)

tmp = df['MaxTemp'].copy()
tmp[df.MaxTemp.isnull()] = mean_MaxTemp
df['MaxTemp_fill'] = tmp
del tmp
df = df.drop(['MaxTemp'],axis=1)

tmp = df['Temp9am'].copy()
tmp[df.Temp9am.isnull()] = mean_Temp9am
df['Temp9am_fill'] = tmp
del tmp
df = df.drop(['Temp9am'],axis=1)

tmp = df['Temp3pm'].copy()
tmp[df.Temp3pm.isnull()] = mean_Temp3pm
df['Temp3pm_fill'] = tmp
del tmp
df = df.drop(['Temp3pm'],axis=1)

tmp = df['Rainfall'].copy()
tmp[df.Rainfall.isnull()] = mean_Rainfall
df['Rainfall_fill'] = tmp
del tmp
df = df.drop(['Rainfall'],axis=1)

tmp = df['Humidity9am'].copy()
tmp[df.Humidity9am.isnull()] = mean_Humidity9am
df['Humidity9am_fill'] = tmp
del tmp
df = df.drop(['Humidity9am'],axis=1)

tmp = df['Humidity3pm'].copy()
tmp[df.Humidity3pm.isnull()] = mean_Humidity3pm
df['Humidity3pm_fill'] = tmp
del tmp
df = df.drop(['Humidity3pm'],axis=1)
```

```python
df.columns  #目前有的變量
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df.groupby(by=['RainToday'])['RainTomorrow'].count() #看raintoday 大部分的答案
```

```python
df.RainToday = [x if str(x) != str('nan') else 'No' for x in df.RainToday] #把raintoday缺失值填為No，不下雨
```

```python
df.groupby(by=['RainToday'])['RainTomorrow'].count() #確認raintoday缺失值填的正確
```

```python
df.isnull().sum()
```

```python
df = df.dropna(axis=0, how='any') #剩下根據有缺失值的刪除其row
```

```python
df.shape #大約刪除10000萬筆 少掉大約7%數據 我認為可以接受
```

```python
df.isnull().sum() #缺失值處理完成
```

**<font size=3 face='STZhongSong'>四、非數值的數據數值化</font>** 

```python
df.head(10) #接下來要將非數值的數據進行數值化
```

```python
df.RainToday = [1 if str(x) == str('Yes') else 0 for x in df.RainToday] #結果是Yes=1;No=0
df.RainTomorrow = [1 if str(x) == str('Yes') else 0 for x in df.RainTomorrow] #結果是Yes=1;No=0
```

```python
df.head(10)
```

```python
df = pd.get_dummies(df) #風的風向進行數值化
df.head(10)
```

**<font size=3 face='STZhongSong'>五、數據建模、訓練、測試、指標比較</font>** 

```python
X = df.loc[:, df.columns!='RainTomorrow'] #第一組
y = df.loc[:, df.columns=='RainTomorrow']
```

```python
from sklearn.model_selection import train_test_split

X_tr, X_ts, y_tr, y_ts = train_test_split(X, y, random_state =2)
X_tr.shape, X_ts.shape
```

```python
from sklearn.naive_bayes import GaussianNB#採用朴素贝叶斯

gnb = GaussianNB()
gnb = gnb.fit(X_tr, y_tr.values.ravel())
```

```python
y_pred = gnb.predict(X_ts)
y_pred[:20]
```

```python
from sklearn.metrics import f1_score
f1_score(y_ts, y_pred)
```

```python
from sklearn.model_selection import cross_val_score#交叉验证
scores = cross_val_score(gnb, X, y.values.ravel(), cv=5, scoring='f1')
scores.mean()
```

```python
#from sklearn.model_selection import cross_val_score

scores_gnb = cross_val_score(gnb, X, y.values.ravel(), cv=5, scoring='accuracy')
scores_gnb.mean()
```

```python
X1 = df.loc[:, df.columns!='RainTomorrow']  #第二組
y1 = df.loc[:, df.columns=='RainTomorrow']
```

```python
from sklearn.model_selection import train_test_split

X1_tr,X1_ts,y1_tr,y1_ts = train_test_split(X1,y1)
X1_tr.shape,X1_ts.shape
```

```python
from sklearn.tree import DecisionTreeClassifier #採用決策樹

dtc = DecisionTreeClassifier()
dtc = dtc.fit(X1_tr,y1_tr)
```

```python
y1_pred = dtc.predict(X1_ts)
y1_pred
```

```python
#from sklearn.metrics import f1_score
f1_score(y1_ts, y1_pred)
```

```python
#from sklearn.model_selection import cross_val_score#交叉验证
scores = cross_val_score(dtc, X1, y1.values.ravel(), cv=5, scoring='f1')
scores.mean()
```

```python
#from sklearn.model_selection import cross_val_score

scores_dtc = cross_val_score(dtc, X1, y1, cv=5, scoring='accuracy')
scores_dtc.mean()
```

```python
X2 = df.loc[:, df.columns!='RainTomorrow']  #第三組
y2 = df.loc[:, df.columns=='RainTomorrow']
```

```python
from sklearn.model_selection import train_test_split

X2_tr,X2_ts,y2_tr,y2_ts = train_test_split(X2,y2)
X2_tr.shape,X2_ts.shape
```

```python
from sklearn.ensemble import RandomForestClassifier  #採用隨機森林
rfc = RandomForestClassifier(n_estimators=10)
rfc = rfc.fit(X2_tr,y2_tr.values.ravel())
```

```python
y2_pred = rfc.predict(X2_ts)
y2_pred
```

```python
#from sklearn.metrics import f1_score
f1_score(y2_ts, y2_pred)
```

```python
#from sklearn.model_selection import cross_val_score#交叉验证
scores = cross_val_score(rfc, X2, y2.values.ravel(), cv=5, scoring='f1')
scores.mean()
```

```python
#from sklearn.model_selection import cross_val_score

scores_dtc = cross_val_score(rfc, X2, y2.values.ravel(), cv=5, scoring='accuracy')
scores_dtc.mean()
```

```python
rfc_cv = cross_val_score(rfc,X2_tr,y2_tr.values.ravel(),scoring='accuracy',cv=5) #should same as above
rfc_cv.mean()
```

```python
from sklearn.model_selection import GridSearchCV, KFold, train_test_split   #採用決策樹格搜索
from sklearn.metrics import make_scorer, accuracy_score
```

```python
regressor = DecisionTreeClassifier(random_state=0)
parameters = {'max_depth': range(1, 6)}
scoring_fnc = make_scorer(accuracy_score)
kfold = KFold(n_splits=10)
```

```python
grid = GridSearchCV(regressor, parameters, scoring_fnc, cv=kfold)
grid = grid.fit(X1_tr, y1_tr)
grid.best_estimator_, grid.best_score_
```

```python
dtc_best = grid.best_estimator_
dtc_best.fit(X1_tr,y1_tr)

y1_pred = dtc_best.predict(X1_ts)
y1_pred[0:5]
```

```python
scores = cross_val_score(dtc_best,X1,y1,cv=5,scoring='accuracy')   #相比之前單純採用決策樹，accuracy比之前0.763提升不少
scores.mean()
```

```python
regressor =RandomForestClassifier(n_estimators=10)       #採用隨機森林格搜索
parameters = {'max_depth': range(1, 6),'min_samples_split':list(range(2,5,1))}
scoring_fnc = make_scorer(accuracy_score)
kfold = KFold(n_splits=10)   
```

```python
grid = GridSearchCV(regressor, parameters, scoring_fnc, cv=kfold)
grid = grid.fit(X2_tr, y2_tr.values.ravel())
grid.best_estimator_, grid.best_score_
```

```python
rfc_best = grid.best_estimator_
rfc_best.fit(X2_tr,y2_tr.values.ravel())

y2_pred = dtc_best.predict(X2_ts)
y2_pred[0:5]
```

```python
scores = cross_val_score(rfc_best,X2,y2.values.ravel(),cv=5,scoring='accuracy')#與單純採用隨機森林，accuracy和之前0.838沒太大改變
scores.mean()
```
