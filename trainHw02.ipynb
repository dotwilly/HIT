**<font size=3 face='STZhongSong'>二、讀取文件</font>** 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from zipfile import ZipFile
df = pd.read_csv('train.csv', index_col=0)
df.shape
```

```python
df.head(5)
```

```python
df.describe()
```

```python
df.isnull().sum().sort_values() #查看缺失值
```

```python
df = df.drop(columns=['LotFrontage','FireplaceQu','Fence','Alley','MiscFeature','PoolQC']) 
#資料缺失千筆的變量刪除，因為總數據也才千筆
df.head(5)
```

```python
df = df.dropna(axis=0, how='any') #剩下根據有缺失值的刪除其row 
```

```python
df.shape   #大約刪除122筆數據 少掉大約8%數據 我認為可以接受
```

```python
df.isnull().sum().sum() #所有資料無缺失值
```

```python
df.describe() #有38筆數據非數值類型
```

```python
df = pd.get_dummies(df) #非數值類型進行數值化
df.head(5)
```

**<font size=3 face='STZhongSong'>三、數據建模、訓練、測試、指標比較</font>** 

```python
X = df.loc[:, df.columns!='SalePrice'] #第一組
y = df.loc[:, df.columns=='SalePrice']
```

```python
X.head(5)
```

```python
y.head(5)
```

```python
import statsmodels.api as sm     #採用多元線性回歸
```

```python
model = sm.OLS(y, X).fit()
model.summary() 
```

```python
y_hat = model.predict(X)
y_hat.head(5)
```

```python
y.shape
```

```python
y_hat.shape
```

```python
y=y.SalePrice
```

```python
y.shape
```

```python
model_RMSE = np.sqrt(np.mean(np.square(y_hat-y)))   #RMSE指標 結果一
```

**<font size=3 face='STZhongSong'>四、總結，預測房價與實際價格大約落在兩萬塊上下，其實滿準的，為方便查看，下面將對SalePrice
進行歸一化</font>** 

```python
max_SalePrice = df['SalePrice'].max()
min_SalePrice = df['SalePrice'].min()
duration=max_SalePrice-min_SalePrice
duration
```

```python
max_SalePrice
```

```python
min_SalePrice
```

```python
df['SalePrice']=df['SalePrice']-min_SalePrice #歸一化
```

```python
df['SalePrice']=df['SalePrice']/duration
df['SalePrice'].head(5)
```

```python
df.shape
```

```python
X = df.loc[:, df.columns!='SalePrice'] #第一組
y = df.loc[:, df.columns=='SalePrice']
```

```python
model = sm.OLS(y, X).fit()
model.summary() 
```

```python
y_hat = model.predict(X)
y_hat.head(5)
```

```python
y_hat.shape
```

```python
y.head(5)
```

```python
y.shape
```

```python
y=y.SalePrice
```

```python
y.shape
```

```python
model_RMSE = np.sqrt(np.mean(np.square(y_hat-y)))    #RMSE指標 結果二 結果挺好的
model_RMSE
```
